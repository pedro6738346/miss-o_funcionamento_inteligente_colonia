```python
#Dados da Colônia

colonia = {
    "Sistema Energético": ["Solar", "Eólico"],
    "Sistema de Alimentação": ["Câmara de Plantação", "Sistema de Irrigação", "Câmara de Sementes"],
    "Sistema Médico": ["Sistema de Atendimento Emergencial", "Sistema de Atendimento Geral"]
}

dados_operacionais = {
    "Energia Atual": 60,
    "Consumo Atual": 80,
    "Vento": 15,
    "Clima": "Nublado"
}

historico_do_vento = [20, 15, 8, 12, 7]
energia_gerada =     [30, 20, 17, 19, 15]

soma_vento = 0
soma_energia = 0

for n in historico_do_vento:
    soma_vento += n

for n in energia_gerada:
    soma_energia += n

media_vento = soma_vento / len(historico_do_vento)
media_energia = soma_energia / len(energia_gerada)
```

```python
#Sistema de Classificação de Dados

def classificacao_dados(dados):
    suporte_vida = True

    if dados["Energia Atual"] < dados["Consumo Atual"]:
        resultado = "ATIVAR RESERVA DE EMERGÊNCIA E LIGAR PAINÉIS SOLARES"

    elif dados["Energia Atual"] < 50 and dados["Consumo Atual"] > 70:
        resultado = "ALERTA! LIGAR MODO ECONOMIA DE ENERGIA"

    else:
        resultado = "CONSUMO DE ENERGIA E NÍVEL DE ENERGIA ACEITÁVEL"

    if suporte_vida:
        return f"SITUAÇÃO: {resultado} | Suporte de Vida: ATIVADO"
    else:
        return "VERIFICAR MÓDULO DE SUPORTE DE VIDA"
```

```python
classificacao_dados(dados_operacionais)
```

```python
#Regressão Linear para Prever a Energia

def prever_energia(vento_atual):
    numerador = 0
    denominador = 0

    for vento, energia in zip(historico_do_vento, energia_gerada):
        numerador += (vento - media_vento) * (energia - media_energia)
        denominador += (vento - media_vento) ** 2

    b = numerador / denominador
    a = media_energia - b * media_vento

    energia_prevista = a + b * vento_atual

    return energia_prevista
```

```python
prever_energia(dados_operacionais['Vento'])
```

```python
geracao = prever_energia(dados_operacionais['Vento'])
consumo = dados_operacionais['Consumo Atual']
```

```python
#Verificar Média de Geração de Energia e Consumo para ver quando Armazenar.

def analisar_energia(geracao, consumo):
    if consumo > geracao:
        resultado = "ALERTA! ATIVAR RESERVA DE ENERGIA!"

    elif geracao > consumo:
        resultado = "ENERGIA EXCEDENTE, ARMAZENAR NA RESERVA DE EMERGÊNCIA"

    else:
        resultado = "NÍVEL DE ENERGIA EQUILIBRADO"

    return resultado
```

```python
analisar_energia(geracao, consumo)
```
